```python
import pandas as pd
from glob import glob
```

```python
stations_files = glob('C:/python/지역*.xls')
stations_files
```

```python
tmp_raw = []
for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
station_raw = pd.concat(tmp_raw)
```

```python
station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame({'011_store':station_raw['상호'],
                         '주소':station_raw['주소'],
                         '가격':station_raw['휘발유'],
                         '셀프':station_raw['셀프여부'],
                         '상표':station_raw['상표']})
stations.head()
```

```python
stations['구'] = [eachAddress.split()[1] for eachAddress in stations['주소']]
stations.head()
```
